# Número de clusters por PCA y método del Codo

```python
# Revisar que las columnas sean numéricas y no tengan valores nulos
df = df.select_dtypes(include=[float, int])

# Estandarizar los datos
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
# Aplicar PCA
pca = PCA()
pca.fit(df_scaled)

plt.figure(figsize=(10, 7))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_.cumsum(), marker='o', linestyle='--')

# Línea del punto del codo
punto_codo = 3
plt.axvline(x=punto_codo, color='r', linestyle='--')

plt.title('Análisis de Componentes Principales')
plt.xlabel('Número de Componentes')
plt.ylabel('Varianza Explicada Acumulativa')
plt.show()
```

En esta gráfica, el punto de codo se encuentra en 3, lo que significa que los tres primeros componentes principales capturan la mayoría de la información significativa de los datos. Hasta el tercer componente, cada componente adicional aporta una cantidad importante de información nueva, lo que se refleja en un fuerte aumento de la varianza explicada acumulativa. Sin embargo, después del tercer componente, los siguientes componentes aportan mucho menos, como se observa por la meseta o el aplanamiento de la curva. Por lo tanto, seleccionar tres componentes maximiza la información obtenida mientras se mantiene la eficiencia al reducir la cantidad de datos, lo que es ideal para simplificar el conjunto de datos sin perder insights valiosos.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np
df_customers = df

# Preprocesamiento: Escalar los datos para mejorar el rendimiento de k-means
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_customers)

# Método del codo para encontrar el número óptimo de clusters
inertia = []
k_values = range(1, 11)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

# Graficar el método del codo
plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia, marker='o')
plt.title('Método del Codo para Determinar el Número Óptimo de Clusters')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.axvline(x=3, color='red', linestyle='--')  # Línea vertical en x = 3
plt.text(3, max(inertia), 'Número óptimo de clusters (3)', color = 'red', ha='center')
plt.grid(True)
plt.show()
```

```python
# Aplicar k-means con el número óptimo de clusters determinado (3 en este caso)
optimal_clusters = 3
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
clusters = kmeans.fit_predict(scaled_features)

# Calcular el puntaje de silueta para evaluar la calidad del clustering
silhouette_avg = silhouette_score(scaled_features, clusters)

# Visualizar los resultados con una gráfica de dispersión
# Asignar las etiquetas de clusters al dataframe original para la visualización
df_customers['Cluster'] = clusters

# Graficar la dispersión por 'Salario' y 'Limite_TC', coloreando por cluster
plt.figure(figsize=(10, 6))
plt.scatter(df_customers['Salario'], df_customers['Limite_TC'], c=df_customers['Cluster'], cmap='viridis')
plt.title('Distribución de los Clientes en Clusters')
plt.xlabel('Salario')
plt.ylabel('Límite de Tarjeta de Crédito')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
print(f'El promedio del método de la silueta para medir la efectividad del modelo es: {silhouette_avg}')
```

# Informe de Evaluación de Desempeño y Explicación del Modelo

## Evaluación del Desempeño:
Para este proyecto hice en primer lugar una clusterización, y para su evaluación utilicé dos métodos: el método del codo y el método de la silueta. El método del codo nos ayuda a determinar el número de grupos distintos en los que podemos dividir a nuestros clientes. Se busca un punto donde, al aumentar el número de grupos, la ganancia de información o "inercia" no mejora significativamente, lo cual se visualiza como un "codo" en la gráfica. Con base en esta técnica, determiné que dividir a los clientes en tres grupos es lo más adecuado.

Además, para entender qué tan bien se ajusta cada cliente dentro de su grupo y qué tan separados están los grupos entre sí, se usó el puntaje de silueta, que es una medida que va de -1 a 1. Un puntaje de 0.33, como el que obtuvimos, indica una calidad moderada de los grupos. Esto significa que los clientes dentro de cada grupo son relativamente similares entre sí, y los grupos son suficientemente distintos unos de otros, aunque hay margen de mejora.

## Explicación del Modelo y Uso de Datos:
El modelo usado se llama k-means, que es una forma común de agrupar datos en grupos basados en similitudes. K-means es como un método para agrupar elementos de acuerdo con distintas características, como se explica en el siguiente punto.
