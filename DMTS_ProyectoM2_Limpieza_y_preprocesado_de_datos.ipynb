Para llenar valores faltantes se utilizó la siguiente lógica:

1. Para variables numéricas continuas, se usó la media principalmente.
2. Para las categóricas, se utilizó la moda.
3. Para cariables con fechas o con secuencias temporales, decidí usar ffil o bfill, así se usa el valor inmediato anterior o posterior, ya que es probable que el dato faltante se acerque a ese valor.

```python
# Función para llenar valores faltantes o eliminar columnas
def llenar_o_eliminar_valores(df, completitud_minima, metodo_por_columna):
    for columna in metodo_por_columna.keys():
        completitud = df[columna].count() / len(df)  # Calcular la completitud
        if completitud > completitud_minima:
            metodo = metodo_por_columna.get(columna, None)  # Obtener el método del diccionario
            if metodo:
                if metodo == 'mean':
                    df[columna].fillna(df[columna].mean(), inplace=True)  # Llenar con la media
                elif metodo == 'median':
                    df[columna].fillna(df[columna].median(), inplace=True)  # Llenar con la mediana
                elif metodo == 'mode':
                    df[columna].fillna(df[columna].mode()[0], inplace=True)  # Llenar con la moda
                elif metodo == 'ffill':
                    df[columna].fillna(method='ffill', inplace=True)  # Llenar con método ffill (adelante)
                elif metodo == 'bfill':
                    df[columna].fillna(method='bfill', inplace=True)  # Llenar con método bfill (atrás)
        else:
            print(f"Eliminando columna '{columna}' debido a baja completitud")
            df.drop(columna, axis=1, inplace=True)  # Eliminar la columna si la completitud es baja
    return df
```

```python
# Función para ajustar valores atípicos según el método elegido
def manejar_outliers_por_metodo(df, metodo='rango', atipicos_threshold=1.5):
    df_manipulado = df.copy()  # Crear una copia del DataFrame para no modificar el original

    for columna in df_manipulado.columns:
        if df_manipulado[columna].dtype != 'object':
            q1 = df_manipulado[columna].quantile(0.25)
            q3 = df_manipulado[columna].quantile(0.75)
            iqr = q3 - q1
            lower_bound = q1 - atipicos_threshold * iqr # q1 - 2*df[columna].std()
            upper_bound = q3 + atipicos_threshold * iqr # q1 + 2*df[columna].std()

            atipicos_indices = (df_manipulado[columna] < lower_bound) | (df_manipulado[columna] > upper_bound)

            if metodo == 'rango':
                df_manipulado[columna] = df_manipulado[columna].clip(lower_bound, upper_bound)  # Ajustar al rango mínimo-máximo
            elif metodo == 'media':
                df_manipulado[columna] = df_manipulado[columna].mask(atipicos_indices, df_manipulado[columna].mean())  # Reemplazar por la media
            elif metodo == 'mediana':
                df_manipulado[columna] = df_manipulado[columna].mask(atipicos_indices, df_manipulado[columna].median())  # Reemplazar por la mediana

    return df_manipulado
```

**Nota:** A pesar de que el método más adecuado para estas variables sería el uso de dummies, por el tamaño del data frame es imposible debido a la cantidad de columnas que generaría y a lo poco óptimo que sería al momento de correr el código.

**Codificación Por Frecuencia**:
La codificación dummies es adecuada para variables categóricas nominales **no ordinales**. Debido a lo mencionado anteriormente, se optó por la codificación por frecuencia. La codificación por frecuencia es una técnica de procesamiento de características categóricas en la que reemplazamos las categorías por la frecuencia (o proporción) con la que aparecen en el conjunto de datos. Es una forma de transformar información categórica en una forma numérica. Las variables codificadas con este método son: term, emp_title, home_ownership, verification_status, issue_d, loan_status, pymnt_plan, purpose, addr_state, earliest_cr_line, initial_list_status, last_pymnt_d y last_credit_pull_d.

**Codificación LabelEncoder**:
La codificación LabelEncoder asigna un valor numérico único a cada categoría de la variable categórica. Es adecuada para variables categóricas **ordinales** donde las categorías tienen un orden claro. Al usar esta codificación, estamos indicando que hay un orden entre las categorías que puede ser útil para el modelo. Las variables codificadas con este método son: grade y sub_grade.


```python
# Creación de listas con las variables que usarán cada uno de los métodos
columnas_dummies = ['term', 'emp_title', 'home_ownership', 'verification_status', 'issue_d', 'loan_status', 'pymnt_plan', 'purpose', 'addr_state', 'earliest_cr_line', 'initial_list_status', 'last_pymnt_d', 'last_credit_pull_d']
columnas_label = ['grade', 'sub_grade']
```

```python
#Función para codificar por frecuencia en lugar de por Dummies
def codificar_por_frecuencia(df, columnas):
    for columna in columnas:
        frecuencias = df[columna].value_counts(normalize=True)
        df[columna] = df[columna].map(frecuencias)

    return df
```

### Conclusión:
La mayoría de la información en los datos está siendo capturada por el primer componente principal. Esto sugiere que muchas características en el conjunto de datos podrían estar altamente correlacionadas. Si el objetivo es reducir la dimensionalidad, considerar solo los primeros 2 o 3 componentes podría ser adecuado, ya que retendrían casi el 98% de la varianza en los datos. En este caso, se decidió tomar los primeros 3 componentes.


```python
# Función para ver cuántas dimensiones se tienen que tomar
## para llegar al número deseado de explicación por varianza
## También regresa el Data Frame ya transformado
def reduccion_pca(df, porcentaje_min_explicacion):
    # Inicializar y ajustar PCA a los datos
    pca = PCA().fit(df)

    # Inicializar varianza explicada acumulada y contador de componentes
    aux = 0
    n = 0

    # Determinar cuántos componentes se necesitan para alcanzar el porcentaje deseado
    while aux < porcentaje_min_explicacion:
        aux += pca.explained_variance_ratio_[n]
        n += 1

    # Ajustar PCA con el número de componentes determinado
    pca = PCA(n_components=n).fit(df)

    # Devolver número de componentes y datos transformados
    return n, pca.transform(df)
```

```python
# Reducción de los datos por 3 componentes (más del 96 % de varianza explicada)
reduccion_pca(df,porcentaje_min_explicacion = 0.96)
```
